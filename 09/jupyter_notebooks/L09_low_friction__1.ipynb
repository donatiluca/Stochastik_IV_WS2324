### Low friction regime
In this notebook, we show that in the low friction regime the total energy of the system varies in a stochastic and diffusive way.

First of all, we consider the case where the system is isolated, i.e. there is no stochastic noise.

```python
# System parameters
kB    = 0.008314463 # kJ mol-1 K
T     = 300         # K
mass  = 1           # amu
beta  = 1 / kB / T

#F R I C T I O N
gamma1 = 0.002      # ps-1 
gamma2 = 0.2       # ps-1 
gamma3 = 20       # ps-1 

D1     = kB * T / mass / gamma1 # nm2 ps-1
D2     = kB * T / mass / gamma2 # nm2 ps-1
D3     = kB * T / mass / gamma3 # nm2 ps-1
```

The following function defines the BBK Langevin integrator.
The option "thermal_fluctuations" is used to switch on/off the thermal fluctuations.
If thermal_fluctuations = 1, we have the Langevin dynamics.
If thermal_fluctuations = 0, we have a dissipative deterministic dynamics due to the friction, then the energy is not conserved.

```python
# Function to define BBK  Langevin integrator
def langevin_bbk_step(Q, P, M, gamma, beta, der_V, dt, thermal_fluctuations = 1):
    L = 1 / (1 + 0.5 * gamma*dt)
    
    # Deterministic force
    F  =  - der_V(Q)
    
    # Random force 
    R  =  thermal_fluctuations * np.random.normal()
    
    # update p_{n+1/2}
    Phalf = ( 1 - 0.5 * gamma * dt ) * P + 0.5 * F * dt + 0.5 * np.sqrt( 2 / beta * dt * gamma * M ) * R
    
    # update q_{n+1}
    Q  =  Q + Phalf / M * dt
    
    # Recalculate deterministic force (but not the random force)
    F  =  - der_V(Q)
    
    # update p_{n+1}
    P = ( Phalf + 0.5 * F * dt + 0.5 * np.sqrt( 2 / beta * dt * gamma * M ) * R ) / ( 1 + 0.5 * gamma * dt ) 

    return Q, P
```

The next block of instructions defines the potential energy function
$$
V(x) = - (x + 1) ^ 3 + \lambda (x + 1) + 2 \, ,
$$

which is characterized by a well on the left side and a barrier on the right side (see figure below the block). The parameter $\lambda$ defines the height of the barrier.

```python
# Parameter Lambda 
L   =  3

# Initialize the variables x and y for symbolic calculation
x   =  symbols('x')

# Define the potentials Va, Vb and V with sympy
V   =   - (x + 1) ** 3 + L * (x + 1) + 2


# Calculate derivative and second derivative with sympy
gradVx   =  V.diff(x)
grad2Vx2 =  gradVx.diff(x)

# To display sympy functions:
# display(gradVx)

# Convert potential and derivatives in numpy functions
V         =  lambdify((x), V, modules=['numpy'])
gradVx    =  lambdify((x), gradVx, modules=['numpy'])
grad2Vx2  =  lambdify((x), grad2Vx2, modules=['numpy'])



xA     = - np.sqrt( L / 3 ) - 1
xB     = + np.sqrt( L / 3 ) - 1

# Approximation at xA and xB
omegaA = np.sqrt( np.abs(grad2Vx2(xA)) / mass )
kspringA     = omegaA ** 2 * mass



omegaB = np.sqrt( np.abs(grad2Vx2(xB)) / mass )
kspringB     = omegaB ** 2 * mass
```

The next block defines a grid for the $x$ axis (position) and the $v$ axis (velocity).
Then it plots the potential.

```python
# Grid
xbins     = 500   
xmin      = - 4
xmax      = - xmin

xedges    = np.linspace(xmin, xmax, xbins)  # array with x edges
dx        = xedges[1] - xedges[0]
xcenters  = xedges[:-1] + (dx / 2)                # array with x centers
xbins     = xbins - 1

vbins     = 200  
vmin      = - 5
vmax      = - vmin

vedges    = np.linspace(vmin, vmax, vbins)  # array with x edges
dv        = vedges[1] - vedges[0]
vcenters  = vedges[:-1] + (dv / 2)                # array with x centers
vbins     = vbins - 1

xv        = np.meshgrid(xcenters, vcenters)
xvf       = np.array([xv[0].flatten('F'), xv[1].flatten('F')]).T





fig, ax1 = plt.subplots(1, 1, figsize=(16*in2cm, 8*in2cm), facecolor='white')  

ax1.plot(xcenters, V(xcenters), 'k', label = 'Potential') 
ax1.plot(xA, V(xA), 'go', label = 'Reactant well') 

ax1.plot(xB, V(xB), 'ro', label = 'Barrier') 

ax1.set_title('Potential')
ax1.set_xlabel(r'$x$')
ax1.set_ylabel(r'$V(x)$')
ax1.legend()
ax1.set_ylim((-5, 15));
```

With the next block, we consider the case of a particle trapped in the well withouth thermal fluctuations. 
It follows that the particle cannot escape and it loses energy at each oscillation.

```python
## Simulation

# Integrator timestep
dt  =  0.005 # ps
sdt =  np.sqrt(dt)

# Number of timesteps
Nsteps  =  10000

# Initial position, momentum and energy
x0  =  xA + 0.5
p0  =  0
E0  =  p0 ** 2 / 2 / mass + V(x0)

# Arrays for x,y values
x1   =   np.zeros(Nsteps)
v1   =   np.zeros(Nsteps)
x2   =   np.zeros(Nsteps)
v2   =   np.zeros(Nsteps)
x3   =   np.zeros(Nsteps)
v3   =   np.zeros(Nsteps)

x1[0] = x0
v1[0] = p0 / mass

x2[0] = x0
v2[0] = p0 / mass

x3[0] = x0
v3[0] = p0 / mass

for k in tqdm(range(Nsteps-1)):
    x1[k+1], p1 = langevin_bbk_step(x1[k], mass * v1[k], mass, gamma1, beta, gradVx, dt, thermal_fluctuations = 0)
    v1[k+1]     = p1 / mass
    
    x2[k+1], p2 = langevin_bbk_step(x2[k], mass * v2[k], mass, gamma2, beta, gradVx, dt, thermal_fluctuations = 0)
    v2[k+1]     = p2 / mass    

    x3[k+1], p3 = langevin_bbk_step(x3[k], mass * v3[k], mass, gamma3, beta, gradVx, dt, thermal_fluctuations = 0)
    v3[k+1]     = p3 / mass    
    
# Calculate energy barrier
Eb = V(xB) - V(xA)

# Calculate energy of each trajectory
E1 = 0.5 * mass * v1 ** 2 + V(x1)
E2 = 0.5 * mass * v2 ** 2 + V(x2)
E3 = 0.5 * mass * v3 ** 2 + V(x3)

# Create a matrix E using the meshgrid xv
# This is necessary to plot the phase space contour lines
E = 0.5 * mass * xv[1]**2 + V(xv[0])

# Array with time values
tvalues = np.linspace(0, Nsteps-1,Nsteps) * dt
```

The thermal noise prevents the system from losing all energy and allows the particle to pass through the barrier.
We observe that when friction is low, the trajectory in phase space does not differ substantially from the fully deterministic case.
The motion is characterized by rapid and regular oscillations, but the time evolution of the total energy appears stochastic and diffusive.
This means that in the low-friction regime, it is possible to define a probability function for the energy, which is a solution of a diffusion equation similar to the Fokker-Planck equation.
For this reason, the low friction regime is also known as energy diffusion limit.

### Low friction condition
Finally, we check if the condition of low friction

$$
\gamma I(E) \ll k_B T \, ,
$$

is satisfied.

```python
def langevin_bbk_step(Q, P, M, gamma, beta, der_V, dt):
    
    Nreps = Q.shape[0]
    
    # bbk Coefficent
    L = 1 / (1 + 0.5 * gamma*dt)
    
    # Deterministic force
    F  =  - der_V(Q)
    
    # Random force 
    R  =  np.random.normal(0, 1, Nreps)
    
    # update p_{n+1/2}
    Phalf = ( 1 - 0.5 * gamma * dt ) * P + 0.5 * F * dt + 0.5 * np.sqrt( 2 / beta * dt * gamma * M ) * R
    
    # update q_{n+1}
    Q  =  Q + Phalf / M * dt
    
    # Recalculate deterministic force (but not the random force)
    F  =  - der_V(Q)
    
    # update p_{n+1}
    P = ( Phalf + 0.5 * F * dt + 0.5 * np.sqrt( 2 / beta * dt * gamma * M ) * R ) / ( 1 + 0.5 * gamma * dt ) 

    return Q, P
```
